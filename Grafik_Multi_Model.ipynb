```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari file laporan iklim
file_path = "Predictions gabungan.csv"
data = pd.read_csv(file_path)

# Pastikan kolom memiliki nama yang sesuai
times = range(len(data['Tanggal']))
true_value = data['Actual']
lstm_prediction = data['LSTM Predictions']
gwo_lstm_prediction = data['LSTM-GWO Predictions']

# Terapkan moving average untuk smoothing
window_size = 10
true_value_smooth = true_value.rolling(window=window_size).mean()
lstm_prediction_smooth = lstm_prediction.rolling(window=window_size).mean()
gwo_lstm_prediction_smooth = gwo_lstm_prediction.rolling(window=window_size).mean()

# Membuat grafik dengan smoothing
plt.figure(figsize=(14, 8))
plt.plot(times, true_value_smooth, label='Nilai Aktual', color='red', linewidth=2, alpha=0.8)
plt.plot(times, lstm_prediction_smooth, label='LSTM', color='green', linewidth=2, alpha=0.8)
plt.plot(times, gwo_lstm_prediction_smooth, label='GWO-LSTM', color='blue', linewidth=2, alpha=0.8)

# Menambahkan label dan judul
plt.xlabel('Waktu', fontsize=14)
plt.ylabel('Temperatur', fontsize=14)
plt.title('Perbandingan Prediksi LSTM, LSTM-GWO, dan Nilai Aktual', fontsize=16)
plt.legend(loc='upper right', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# Tampilkan grafik
plt.show()
```
