### Handle Missing Value dengan Metode Interpolasi
Alasan menggunakan interpolasi untuk handle missing value karena interpolasi menggunakan informasi dari data yang ada di sekitar missing value, sehingga mengurangi risiko pengenalan bias yang bisa terjadi. Data time series juga bersifat urutan kronologis yang memerlukan kesinambungan untuk analisis yang lebih akurat. Interpolasi mengisi missing values sehingga pola dan tren tidak terputus. Metode interpolasi juga memperhitungkan urutan waktu dan ketergantungan temporal antara data points. Hal ini sangat penting dalam time series karena nilai pada suatu waktu seringkali bergantung pada nilai sebelumnya.

```python
# Menggunakan interpolate() untuk mengisi nilai yang hilang
df['Temperatur_Minimum'] = df['Temperatur_Minimum'].interpolate()
df['Temperatur_Maksimum'] = df['Temperatur_Maksimum'].interpolate()
df['Temperatur_Rata_Rata'] = df['Temperatur_Rata_Rata'].interpolate()
df['Kelembapan_Rata_Rata'] = df['Kelembapan_Rata_Rata'].interpolate()
df['Curah_Hujan'] = df['Curah_Hujan'].interpolate()
df['Lamanya_Penyinaran'] = df['Lamanya_Penyinaran'].interpolate()
df['Kecepatan_Angin_Maksimum'] = df['Kecepatan_Angin_Maksimum'].interpolate()
df['Arah_Angin_Maksimum'] = df['Arah_Angin_Maksimum'].interpolate()
df['Kecepatan_Angin_Rata_Rata'] = df['Kecepatan_Angin_Rata_Rata'].interpolate()
```

Penjelasan Tambahan:

- IQR adalah metode yang umum digunakan untuk mendeteksi outliers.

```python
import pandas as pd
import numpy as np

# Mengganti nilai 8888 dan 9999 dengan NaN terlebih dahulu
df.replace([8888, 9999], np.nan, inplace=True)

# Fungsi untuk mengganti outliers menggunakan interpolasi
def replace_outliers_with_interpolate(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Mengganti outliers dengan interpolasi
    column[(column < lower_bound) | (column > upper_bound)] = np.nan
    column = column.interpolate()  # Menggunakan interpolasi untuk mengisi NaN
    return column

# Menerapkan penggantian outliers pada kolom-kolom yang terdeteksi
for col in ['Temperatur_Minimum', 'Temperatur_Maksimum', 'Temperatur_Rata_Rata', 'Curah_Hujan', 'Kelembapan_Rata_Rata', 'Kecepatan_Angin_Maksimum', 'Kecepatan_Angin_Rata_Rata']:
    df[col] = replace_outliers_with_interpolate(df[col])

# Mengecek hasil
print(df.head())
```
